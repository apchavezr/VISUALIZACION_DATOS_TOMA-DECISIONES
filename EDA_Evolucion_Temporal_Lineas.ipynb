# Visualización de la evolución temporal con gráficos de líneas

Este ejemplo básico ilustra cómo utilizar gráficos de líneas para analizar la evolución de un indicador en el tiempo. Este tipo de representación permite observar tendencias, patrones estacionales o posibles rupturas estructurales en una serie temporal.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Cargar los datos simulados
datos = pd.read_csv("serie_temporal_ejemplo.csv", parse_dates=["Fecha"])
datos.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(datos['Fecha'], datos['Valor'], marker='o', linestyle='-', color='steelblue')
plt.title('Evolución temporal del indicador')
plt.xlabel('Fecha')
plt.ylabel('Valor del indicador')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Interpretación

- El gráfico de líneas muestra una **tendencia creciente** en el tiempo.
- Se puede notar una **variación cíclica** que sugiere **estacionalidad anual**.
- Este tipo de análisis es común en indicadores económicos (como inflación o empleo), métricas de plataformas digitales (como número de usuarios), o monitoreo de políticas públicas.

La visualización temporal es un recurso clave en análisis de series de tiempo, ya que permite interpretar fenómenos que cambian de forma continua en el tiempo.
