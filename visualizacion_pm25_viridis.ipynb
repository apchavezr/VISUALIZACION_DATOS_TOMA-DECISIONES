**Ejemplo práctico**

Considere un mapa de calor que visualiza la calidad del aire en ciudades de Colombia. Para representar los niveles de PM2.5 (material particulado fino), se recomienda una escala continua como Viridis:


*   Verde oscuro: niveles bajos (0-12 µg/m³)
*   Amarillo: niveles moderados (12-35 µg/m³)
*   Púrpura: niveles altos (>35 µg/m³)


```python
# Visualización básica de niveles de PM2.5 en ciudades colombianas usando Viridis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Datos simulados
ciudades = ['Bogotá', 'Medellín', 'Cali', 'Barranquilla', 'Bucaramanga']
np.random.seed(42)
valores_pm25 = np.random.uniform(5, 45, size=5)

df = pd.DataFrame({
    'Ciudad': ciudades,
    'PM2.5': valores_pm25
})

# Visualización tipo heatmap
plt.figure(figsize=(8, 3))
pivot = pd.DataFrame([valores_pm25], columns=ciudades)
sns.heatmap(pivot, cmap='viridis', annot=True, fmt='.1f', cbar_kws={'label': 'PM2.5 (µg/m³)'})
plt.title('Mapa de calor - Niveles de PM2.5 por ciudad')
plt.yticks([])
plt.tight_layout()
plt.show()
```
