# Visualización de datos: uso de atributos preatencionales

Este cuaderno presenta un ejemplo práctico del uso de **atributos preatencionales** en visualización de datos.
Los atributos preatencionales son aquellos rasgos visuales como el **color**, **tamaño**, **forma** u **orientación** que el ojo humano puede identificar **sin necesidad de atención consciente**.

## Objetivo del ejemplo
Demostrar cómo el uso estratégico del color y el tamaño puede destacar visualmente una categoría de datos, guiando la atención del observador hacia un punto específico de interés.

Referencia:
- Cairo, A. (2016). *El arte funcional: una introducción a la infografía y la visualización de la información*. Alamut.


```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Crear un conjunto de datos de ejemplo
df = pd.DataFrame({
    'Categoría': ['A', 'B', 'C', 'D', 'E'],
    'Valor': [10, 15, 8, 25, 12]
})

# 2. Configurar atributos preatencionales
#    - Color: destacamos la categoría 'D' con un color rojo
#    - Tamaño: aumentamos el tamaño del punto correspondiente a 'D'
colors = ['gray' if cat != 'D' else 'crimson' for cat in df['Categoría']]
sizes = [100 if cat != 'D' else 300 for cat in df['Categoría']]

# 3. Crear un gráfico tipo scatter para representar los datos
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(df['Categoría'], df['Valor'], s=sizes, c=colors)

# 4. Etiquetas y títulos para contexto
ax.set_title("Uso de atributos preatencionales en visualización")
ax.set_ylabel("Valor")
ax.set_xlabel("Categoría")

# 5. Añadir etiquetas con los valores encima de cada punto
for i, row in df.iterrows():
    ax.text(row['Categoría'], row['Valor'] + 0.8, str(row['Valor']), ha='center')

# 6. Ajuste final del diseño
plt.tight_layout()
plt.show()
```

## Conclusión

En este ejemplo se observa cómo el uso de color y tamaño, aplicados correctamente, puede:

- Dirigir la atención hacia un dato relevante (en este caso, la categoría 'D').
- Mejorar la interpretación inmediata del gráfico.
- Facilitar la toma de decisiones al resaltar patrones o valores atípicos.

Este enfoque resulta útil en informes, presentaciones y tableros de control que requieren destacar información clave de forma intuitiva.
