# An√°lisis de Distribuci√≥n y Comparaci√≥n entre Grupos con Visualizaci√≥n Gr√°fica

Este cuaderno presenta un ejemplo b√°sico sobre c√≥mo analizar la **distribuci√≥n de una variable** y c√≥mo **comparar esa variable entre grupos categ√≥ricos**, utilizando gr√°ficos como histogramas, diagramas de densidad, gr√°ficos de caja y gr√°ficos de viol√≠n.

Se simulan datos de puntajes en una prueba de comprensi√≥n lectora en dos colegios diferentes, y se explora visualmente la informaci√≥n.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Simulaci√≥n
np.random.seed(10)
n = 100
data = pd.DataFrame({
    'Colegio': ['A'] * (n // 2) + ['B'] * (n // 2),
    'Puntaje': np.concatenate([
        np.random.normal(70, 10, n // 2),  # Colegio A
        np.random.normal(75, 8, n // 2)    # Colegio B
    ])
})

data.head()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data['Puntaje'], kde=True, bins=15, color='lightblue')
plt.title('Distribuci√≥n general de puntajes')
plt.xlabel('Puntaje')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Colegio', y='Puntaje', data=data, palette='pastel')
plt.title('Distribuci√≥n de puntajes por colegio (Boxplot)')
plt.xlabel('Colegio')
plt.ylabel('Puntaje')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x='Colegio', y='Puntaje', data=data, palette='muted')
plt.title('Distribuci√≥n de puntajes por colegio (Gr√°fico de viol√≠n)')
plt.xlabel('Colegio')
plt.ylabel('Puntaje')
plt.show()
```

## Conclusiones

- El **histograma con densidad** permite observar la forma general de la distribuci√≥n de los puntajes, incluyendo concentraci√≥n y simetr√≠a.
- El **boxplot** muestra la mediana, rango intercuart√≠lico y valores at√≠picos para cada colegio.
- El **gr√°fico de viol√≠n** permite observar visualmente la forma completa de la distribuci√≥n en cada grupo, resaltando diferencias en densidad y dispersi√≥n.
- Este tipo de visualizaci√≥n es √∫til en an√°lisis educativos y comparativos, ayudando a orientar decisiones antes de aplicar t√©cnicas estad√≠sticas formales.
