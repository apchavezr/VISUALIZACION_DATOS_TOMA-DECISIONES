# Análisis Visual de Correlaciones entre Variables Cuantitativas

Este notebook presenta un ejemplo básico para ilustrar el uso de **diagramas de dispersión**, **matrices de correlación** y **gráficos de pares** (pairplots) en el análisis exploratorio de datos (EDA). Estas herramientas permiten estudiar visualmente relaciones lineales o no lineales entre variables numéricas.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
sns.set(style="whitegrid")
np.random.seed(42)

# Simulación de un conjunto de datos con correlaciones
n = 100
x = np.random.normal(loc=50, scale=10, size=n)
y = x * 1.5 + np.random.normal(0, 5, n)
z = x * -0.8 + np.random.normal(0, 8, n)

df = pd.DataFrame({
    'Ingreso': x,
    'Gasto': y,
    'Ahorro': z
})

df.head()
```

## Diagrama de dispersión

El siguiente gráfico muestra la relación entre dos variables: Ingreso y Gasto.


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Ingreso', y='Gasto', data=df, color='teal')
plt.title('Relación entre Ingreso y Gasto')
plt.xlabel('Ingreso mensual (en miles)')
plt.ylabel('Gasto mensual (en miles)')
plt.show()
```

## Matriz de correlación

La matriz de correlación muestra el coeficiente de correlación de Pearson entre todas las variables cuantitativas del dataset.


```python
plt.figure(figsize=(6, 4))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de correlación entre variables')
plt.show()
```

## Gráfico de pares (pairplot)

El gráfico de pares permite visualizar todas las relaciones bivariadas posibles en un conjunto de variables numéricas.


```python
sns.pairplot(df)
plt.suptitle('Visualización de relaciones bivariadas', y=1.02)
plt.show()
```

## Conclusiones

- El **diagrama de dispersión** muestra una relación lineal positiva fuerte entre Ingreso y Gasto.
- La **matriz de correlación** confirma esta relación (coeficiente cercano a 1), e indica una correlación negativa entre Ingreso y Ahorro.
- El **gráfico de pares** permite observar estas relaciones de manera simultánea y detectar patrones, tendencias o posibles agrupamientos.

Este tipo de análisis es útil en estudios financieros, encuestas socioeconómicas o análisis de consumo.
