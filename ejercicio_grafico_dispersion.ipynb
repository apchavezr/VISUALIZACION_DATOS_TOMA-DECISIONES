# Gráfico de dispersión: Horas de estudio vs Puntaje en el examen

Este ejercicio demuestra cómo construir e interpretar un gráfico de dispersión usando Python. Se analiza la relación entre las horas dedicadas al estudio y el puntaje obtenido en un examen.



````python
# Importar librerías necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Simulación de datos bivariados
np.random.seed(42)
x = np.random.uniform(10, 100, 50)  # Horas de estudio
y = 2.5 * x + np.random.normal(0, 20, 50)  # Puntaje en el examen con ruido

# Crear DataFrame
df = pd.DataFrame({
    'Horas de estudio': x,
    'Puntaje en el examen': y
})

# Visualización del gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Horas de estudio', y='Puntaje en el examen', color='blue', s=80)
sns.regplot(data=df, x='Horas de estudio', y='Puntaje en el examen', scatter=False, color='red')

# Personalización
plt.title('Gráfico de dispersión: Horas de estudio vs Puntaje en el examen')
plt.xlabel('Horas de estudio')
plt.ylabel('Puntaje en el examen')
plt.grid(True)
plt.tight_layout()
plt.show()
````

> 🧠 **Reflexión**:
>
*  **¿Observa una tendencia positiva o negativa?**
>Se observa una tendencia positiva. A medida que aumentan las horas de estudio, el puntaje en el examen también tiende a aumentar. Esta relación ascendente es visible en la alineación general de los puntos en dirección de izquierda a derecha y de abajo hacia arriba, y se refuerza por la pendiente positiva de la línea de regresión.

*  **¿Qué tan fuerte parece la relación entre horas de estudio y puntaje?**
>La relación es moderadamente fuerte. Aunque muchos puntos se agrupan alrededor de la línea de tendencia, también hay una dispersión considerable debido al ruido aleatorio incorporado en los datos. Esto sugiere una correlación positiva clara, pero no perfecta. En un contexto real, esto podría interpretarse como que estudiar más generalmente mejora el desempeño, aunque otros factores también influyen.

*  **¿Se identifican valores atípicos?**
> Sí, se pueden identificar algunos valores atípicos. Por ejemplo, hay observaciones con muchas horas de estudio pero con un puntaje relativamente bajo (por debajo de la línea de tendencia), lo cual puede deberse a errores, condiciones externas o casos atípicos que ameritan análisis adicional. Estos puntos no siguen el patrón general y se ubican más alejados de la línea de tendencia.
